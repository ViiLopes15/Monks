# Imports




```python
import pandas as pd
import requests
import os
```

# Requisição Token de Acessso

# Variáveis de ambiente

```python
client_id = ('ae77092415e944ffab7dc85263ef6fa2')
client_secret = ('5b1f7285097b4cc69e08e38f278c1fb5')
```

# Conversão da Autenticação para Base 64

```python
import base64

string = client_id + ':' + client_secret
string_bytes = string.encode('ascii')

base64_bytes = base64.b64encode(string_bytes)
base64_string = base64_bytes.decode('ascii')
```

# Requisição


```python
url = 'https://accounts.spotify.com/api/token'

headers = {'Authorization': f'Basic {base64_string}',
           'Content-Type': 'application/x-www-form-urlencoded'}

payload = {'grant_type': 'client_credentials'}

response = requests.request('POST', url = url, headers = headers, data = payload)

```

```python
response.json()
```

```python
access_token = response.json()['access_token']
```

# Exemplo de requisição

```python
url = 'https://api.spotify.com/v1/tracks/0tgVpDi06FyKpA1z0VMD4v'

headers = {'Authorization': f'Bearer {access_token}'}

response = requests.request('GET', url = url, headers = headers)

response.status_code
```

```python
response.json()
```
